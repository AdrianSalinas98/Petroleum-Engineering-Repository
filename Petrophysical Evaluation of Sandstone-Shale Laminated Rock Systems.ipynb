```python
#This Workflow was made to evaluate the composition and storage properties of sandstone with laminated shale. 

#It will return: Concentration of Shale, Sandstone Porosity, Water Saturation, Hydrocarbon Saturation, 
               # Hydrocarbon Pore Volume per unit depth, Fluid Density, Hydrocarbon Density, and Fluid Type

# case 1: If clean, water saturated sandstone and pure shale available

# Case 2: If formation of interest is shale-laminated water saturated Sandstone(Aquifer): d/D
    
# Case 3: If formation of interest is not water saturated
        # If gamma ray not available, perpendicular and parallel resistivity available: 
        # solve system for Csh & Rss

#Case 4: If reference Sandstone water saturated(Aquifer) either clean or laminated, 
       # and both perpendicular and parallel resistivity not available.

#workflow is best used with Well-Log data with the following tracks: Gamma Ray, Resistivity, Porosity
# assumes that all sandstones and shales are of the same composition across pure and laminated sections, hydraulic
# communication between aquifer and shallower zones with water


request = int(input('Enter the Case Number: '))


# case 1: If clean, water saturated sandstone and pure shale available

# shale concentration
if request == 1:
    clean_sand_gamma = float(input('Give Gamma Ray API value for clean sand: '))
    pure_shale_gamma = float(input('Give Gamma Ray API value for pure shale: '))

    laminated_formation = float(input('Give Gamma Ray API value for the laminated formation: '))
    
    Csh = (laminated_formation - clean_sand_gamma)/(pure_shale_gamma-clean_sand_gamma)

#__________________________________________________________
# sandstone porosity
    neutron_reference = float(input('Give Neutron value in SS porosity units Laminated Formation: '))
    density_reference = float(input('Give Density value in SS porosity units Laminated Formation: '))
    neutron_shale = float(input('Give Neutron value in SS porosity units across a pure shale: '))
    density_shale = float(input('Give Density value in SS porosity units across a pure shale: '))
    neutron_Aq = float(input('Give Neutron value in SS porosity units Aquifer: '))
    density_Aq = float(input('Give Density value in SS porosity units Aquifer: '))
    
   
    SSporosity_Aq = ((((density_Aq**2) + (neutron_Aq**2))/2)**.5)
    
    
    correctedDensityPorosity = (density_reference-(Csh*density_shale))/(1-Csh)
    correctedNuetronPorosity = (neutron_reference-(Csh*neutron_shale))/(1-Csh)
    
    correctedSSporosity = (((correctedDensityPorosity**2) + (correctedNuetronPorosity**2))/2)**.5


#__________________________________________________________
# Saturations

    horizontalResistivity = float(input('Give deep horizontal resistivity of laminated formation in Ohm-m: '))
    Rss_Aq = float(input('Give deep horizontal resistivity of the Aquifer(clean sand) formation in Ohm-m: '))
    shaleResistivity = float(input('Give deep resistivity of pure shale in Ohm-m: '))


    
    Rss = (1-Csh)/((1/horizontalResistivity)-(Csh/shaleResistivity)) #resistivity of sand in laminated formation

    m = float(input("Give Archie's 'm' parameter: "))
    n = float(input("Give Archie's 'n' parameter: "))
    a = float(input("Give Archie's 'a' parameter: "))
    
    
    Rw = (Rss_Aq*((SSporosity_Aq**m)/a)) #resisitivity of formation water in sand in laminated formation
    Sw = ((Rw/Rss)*(a/(correctedSSporosity**m)))**(1/n)#water saturation in sand in laminated formation
    Shc = (1-Sw)
    
    HPV = (1-Csh)*(correctedSSporosity)*(Shc) #hydrocarbon Pore Volume per unit depth
#__________________________________________________________
# Fluid Composition
    
    rhoB = (density_reference*1) + (1-density_reference)*2.65
    rhoSh = (density_shale*1) + (1-density_shale)*2.65
    rhoSS = (1/(1-Csh))*(rhoB-(rhoSh*Csh))
    rhoF = (1/correctedSSporosity)*(rhoSS-(1-correctedSSporosity)*2.65)
    
    if Sw == 1:
        rhoHC = 'Aquifer'
    else:
        rhoHC = (1/(1-Sw))*(rhoF-(Sw*1))
    
    if rhoHC == 'Aquifer':
        fluid = 'Water'
    elif rhoHC < 0.3:
        fluid = 'light hydrocarbon(need context)' 
    else:
        fluid = 'oil(need context)'     
              
#__________________________________________________________________________________________________________________    
    
# Case 2: If formation of interest is shale-laminated water saturated Sandstone(Aquifer): d/D


# shale concentration
elif request == 2:
    neutron_reference = float(input('Give Neutron value in SS porosity units across aquifer: '))
    density_reference = float(input('Give Density value in SS porosity units across aquifer: '))
    neutron_shale = float(input('Give Neutron value in SS porosity units across a pure shale: '))
    density_shale = float(input('Give Density value in SS porosity units across a pure shale: '))
    
    Csh = (neutron_reference-density_reference) / (neutron_shale-density_shale)
        
    correctedDensityPorosity = (density_reference-(Csh*density_shale))/(1-Csh)
    correctedNuetronPorosity = (neutron_reference-(Csh*neutron_shale))/(1-Csh)
    
    correctedSSporosity = (((correctedDensityPorosity**2) + (correctedNuetronPorosity**2))/2)**.5
    
#__________________________________________________________
# sandstone porosity
    
    correctedDensityPorosity = (density_reference-(Csh*density_shale))/(1-Csh)
    correctedNuetronPorosity = (neutron_reference-(Csh*neutron_shale))/(1-Csh)
    
    correctedSSporosity = (((correctedDensityPorosity**2) + (correctedNuetronPorosity**2))/2)**.5
#__________________________________________________________
# Saturations

    horizontalResistivity = float(input('Give deep horizontal resistivity of laminated formation in Ohm-m: '))
    shaleResistivity = float(input('Give deep resistivity of pure shale in Ohm-m: '))
    
    Rss = (1-Csh)/((1/horizontalResistivity)-(Csh/shaleResistivity)) #resistivity of sand in laminated formation
    m = float(input("Give Archie's 'm' parameter: "))
    n = float(input("Give Archie's 'n' parameter: "))
    a = float(input("Give Archie's 'a' parameter: "))

    Rw = (Rss*((correctedSSporosity**m)/a)) #resisitivity of formation water in sand in laminated formation
    Sw = ((Rw/Rss)*(a/(correctedSSporosity**m)))**(1/n)#water saturation in sand in laminated formation
    Shc = (1-Sw)
    
    HPV = (1-Csh)*(correctedSSporosity)*(Shc) #hydrocarbon Pore Volume per unit depth
#__________________________________________________________
# Fluid Composition
    
    rhoB = (density_reference*1) + (1-density_reference)*2.65
    rhoSh = (density_shale*1) + (1-density_shale)*2.65
    rhoSS = (1/(1-Csh))*(rhoB-(rhoSh*Csh))
    rhoF = (1/correctedSSporosity)*(rhoSS-(1-correctedSSporosity)*2.65)
    
    if Sw == 1:
        rhoHC = 'Aquifer'
    else:
        rhoHC = (1/(1-Sw))*(rhoF-(Sw*1))

    if rhoHC == 'Aquifer':
        fluid = 'Water'
    elif rhoHC < 0.3:
        fluid = 'light hydrocarbon(need context)' 
    else:
        fluid = 'oil(need context)'
    
#__________________________________________________________________________________________________________________    

        
# Case 3: if formation of interest is not water saturated
        # and If gamma ray not available and, perpendicular and parallel resistivity available: 
        # solve system for Csh & Rss
        
elif request == 3:
# shale concentration
    neutron_reference = float(input('Give Neutron value in SS porosity units across lanimated formation: '))
    density_reference = float(input('Give Density value in SS porosity units across laminated formation: '))
    neutron_shale = float(input('Give Neutron value in SS porosity units across a pure shale: '))
    density_shale = float(input('Give Density value in SS porosity units across a pure shale: '))
    
    horizontalResistivity = float(input('Give deep horizontal resistivity of laminated formation in Ohm-m: '))
    verticalResistivity = float(input('Give vertical resistivity of laminated formation in Ohm-m: '))
    shaleResistivity = float(input('Give deep resistivity of pure shale in Ohm-m: '))

    Csh = ((horizontalResistivity*shaleResistivity)-(verticalResistivity*shaleResistivity))/((
        2*horizontalResistivity*shaleResistivity)-(shaleResistivity**2)-(verticalResistivity*horizontalResistivity))
#__________________________________________________________
# sandstone porosity   
    correctedDensityPorosity = (density_reference-(Csh*density_shale))/(1-Csh)
    correctedNuetronPorosity = (neutron_reference-(Csh*neutron_shale))/(1-Csh)
    
    correctedSSporosity = (((correctedDensityPorosity**2) + (correctedNuetronPorosity**2))/2)**.5

#__________________________________________________________
# Saturations

    neutron_reference_Aq = float(input('Give Neutron value in SS porosity units across Aquifer: '))
    density_reference_Aq = float(input('Give Density value in SS porosity units across Aquifer: '))
    Csh_Aq = (neutron_reference_Aq-density_reference_Aq) / (neutron_shale-density_shale)



    
    Rss = (1-Csh)/((1/horizontalResistivity)-(Csh/shaleResistivity))

    m = float(input("Give Archie's 'm' parameter: "))
    n = float(input("Give Archie's 'n' parameter: "))
    a = float(input("Give Archie's 'a' parameter: "))
    
    horizontalResistivity_Aq = float(input('Give deep horizontal resistivity of Aquifer in Ohm-m: '))
    Rss_Aq = (1-Csh_Aq)/((1/horizontalResistivity_Aq)-(Csh_Aq/shaleResistivity))


    correctedNuetronPorosityAq = (neutron_reference_Aq - Csh_Aq*neutron_shale)/(1-Csh_Aq)
    correctedDensityPorosityAq = (density_reference_Aq - Csh_Aq*density_shale)/(1-Csh_Aq)
    correctedSSporosity_Aq = ((((correctedDensityPorosityAq**2) + (correctedNuetronPorosityAq**2))/2)**.5)

    
    Rw = (Rss_Aq*((correctedSSporosity_Aq**m)/a)) #resisitivity of formation water in sand in laminated formation
    Sw = ((Rw/Rss)*(a/(correctedSSporosity**m)))**(1/n)#water saturation in sand in laminated formation
    Shc = (1-Sw)
    
    HPV = (1-Csh)*(correctedSSporosity)*(Shc) #hydrocarbon Pore Volume per unit depth
    
#__________________________________________________________
# Fluid Composition
    
    rhoB = (density_reference*1) + (1-density_reference)*2.65
    rhoSh = (density_shale*1) + (1-density_shale)*2.65
    rhoSS = (1/(1-Csh))*(rhoB-(rhoSh*Csh))
    rhoF = (1/correctedSSporosity)*(rhoSS-(1-correctedSSporosity)*2.65)


    if Sw == 1:
        rhoHC = 'Aquifer'
    else:
        rhoHC = (1/(1-Sw))*(rhoF-(Sw*1))
    
    if rhoHC == 'Aquifer':
        fluid = 'Water'
    elif rhoHC < 0.3:
        fluid = 'light hydrocarbon(need context)' 
    else:
        fluid = 'oil(need context)' 
        
#__________________________________________________________________________________________________________________    

#Case 4: If reference Sandstone water saturated(Aquifer) either clean or laminated, 
       # and both perpendicular and parallel resistivity not available.

elif request == 4:
    neutron_reference = float(input('Give Neutron value in SS porosity units across Laminated formation: '))
    density_reference = float(input('Give Density value in SS porosity units across Laminated formation: '))
    neutron_shale = float(input('Give Neutron value in SS porosity units across a pure shale: '))
    density_shale = float(input('Give Density value in SS porosity units across a pure shale: '))
    neutron_Aq = float(input('Give Neutron value in SS porosity units Aquifer: '))
    density_Aq = float(input('Give Density value in SS porosity units Aquifer: '))
    
    reference_aquifer_gamma = float(input('Give Gamma Ray API value for Aquifer: '))
    pure_shale_gamma = float(input('Give Gamma Ray API value for pure shale: '))
    
    Csh_Aq = (neutron_Aq-density_Aq) / (neutron_shale-density_shale)
    slope = (1-Csh_Aq)/(pure_shale_gamma-reference_aquifer_gamma)
    
    formation_of_interest_gamma = float(input('Give Gamma Ray API value for formation of interest: '))
    
    Csh = slope*(formation_of_interest_gamma-pure_shale_gamma)+ 1
    
#__________________________________________________________
# sandstone porosity   

    correctedDensityPorosity = (density_reference-(Csh*density_shale))/(1-Csh)
    correctedNuetronPorosity = (neutron_reference-(Csh*neutron_shale))/(1-Csh)
    correctedSSporosity = (((correctedDensityPorosity**2) + (correctedNuetronPorosity**2))/2)**.5
    
    correctedNuetronPorosityAq = (neutron_Aq - Csh_Aq*density_shale)/(1-Csh_Aq)
    correctedDensityPorosityAq = (density_Aq - Csh_Aq*neutron_shale)/(1-Csh_Aq)
    correctedSSporosity_Aq = ((((correctedDensityPorosityAq**2) + (correctedNuetronPorosityAq**2))/2)**.5)

#__________________________________________________________
# Saturations

    horizontalResistivity = float(input('Give deep horizontal resistivity of laminated formation in Ohm-m: '))
    shaleResistivity = float(input('Give deep resistivity of pure shale in Ohm-m: '))
    
    Rss = (1-Csh)/((1/horizontalResistivity)-(Csh/shaleResistivity))
    print(Rss,'Rss')
    m = float(input("Give Archie's 'm' parameter: "))
    n = float(input("Give Archie's 'n' parameter: "))
    a = float(input("Give Archie's 'a' parameter: "))
    
    horizontalResistivity_Aq = float(input('Give deep horizontal resistivity of Aquifer in Ohm-m: '))
    Rss_Aq = (1-Csh_Aq)/((1/horizontalResistivity_Aq)-(Csh_Aq/shaleResistivity))
    
    
    
    
    Rw = (Rss_Aq*((correctedSSporosity_Aq**m)/a)) #resisitivity of formation water in sand in laminated formation
    print(Rw,'Rw')
    Sw = ((Rw/Rss)*(a/(correctedSSporosity**m)))**(1/n)#water saturation in sand in laminated formation
    Shc = (1-Sw)
    
    HPV = (1-Csh)*(correctedSSporosity)*(Shc) #hydrocarbon Pore Volume per unit depth
    
#__________________________________________________________
# Fluid Composition
    
    rhoB = (density_reference*1) + (1-density_reference)*2.65
    rhoSh = (density_shale*1) + (1-density_shale)*2.65
    rhoSS = (1/(1-Csh))*(rhoB-(rhoSh*Csh))
    rhoF = (1/correctedSSporosity)*(rhoSS-(1-correctedSSporosity)*2.65)
    
    if Sw == 1:
        rhoHC = 'Aquifer'
    else:
        rhoHC = (1/(1-Sw))*(rhoF-(Sw*1))
    
    if rhoHC == 'Aquifer':
        fluid = 'Water'
    elif rhoHC < 0.3:
        fluid = 'light hydrocarbon(need context)' 
    else:
        fluid = 'oil(need context)'   

print()
print('-'*60)
print('Results: ')
print('-'*60)
print('   Concentration of Shale:'+ format(Csh*100,'>30.2f') +'%')
print('   Sandstone Porosity:' + format(correctedSSporosity*100,'>34.2f')+'%')
print('   Water Saturation:' + format(Sw*100,'>36.2f')+'%')
print('   Hydrocarbon Saturation:' + format(Shc*100,'>30.2f')+'%')
print('   Hydrocarbon Pore Volume per unit depth: ' + format(HPV,'>14.2f'))
print('   Fluid Density:' + format(rhoF,'>35.2f')+' g/cc')

if Sw == 1:
    print('   Hydrocarbon Density:' + format(rhoHC,'>34'))
else: 
    print('   Hydrocarbon Density:' + format(rhoHC,'>29.2f')+' g/cc')
if Sw == 1:
    print('   Likely Fluid Type:' + format('water','>43'))
else: 
    print('   Likely Fluid Type:' + format(fluid,'>36'))
print('-'*60)


```
